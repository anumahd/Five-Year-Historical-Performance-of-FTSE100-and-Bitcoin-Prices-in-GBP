##### To conduct a hypothesis test to determine whether there is a significant difference between the price movements of FTSE100 and Bitcoin, we'll use a two-sample t-test. This test compares the means of two independent samples (in this case, the daily returns of FTSE100 and Bitcoin) to see if they are significantly different from each other.

```python
#Steps
#Calculate daily returns for both FTSE100 and Bitcoin.
#Check assumptions:
#Normality: The returns should be approximately normally distributed.
#Homogeneity of variances: The variances of the returns should be approximately equal.
#Conduct the two-sample t-test.
#Interpret the results: Compare the p-value to the significance level (0.05) to decide whether to reject the null hypothesis.
```

```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
import scipy.stats as stats
import matplotlib.pyplot as plt

# same data would be used from the already processed and extracted data set (df)

df = pd.read_csv("final_data_AXA.csv")
```

```python
# Conduct the two-sample t-test
t_stat, p_value = stats.ttest_ind(df['FTSE100_Return'], df['Bitcoin_Return'], equal_var=False)
```

```python
# Print the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the daily returns of FTSE100 and Bitcoin.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the daily returns of FTSE100 and Bitcoin.")
```

```python

# Visualizing the returns
plt.figure(figsize=(16, 8))
plt.hist(df['FTSE100_Return'], bins=50, alpha=0.5, label='FTSE100 Returns', color='blue')
plt.hist(df['Bitcoin_Return'], bins=50, alpha=0.5, label='Bitcoin Returns', color='orange')
plt.title('Distribution of Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
